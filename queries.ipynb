# Importando Bibliotecas

```python
import pandas as pd
from google.cloud import bigquery
from google.oauth2 import service_account
```

# 4. Utilizando consultas SQL responda as perguntas

```python
credentials = service_account.Credentials.from_service_account_file('svc-data-engineer-test.json', scopes =["https://www.googleapis.com/auth/cloud-platform"])
```

# 4.1 - Quantos usuários temos por estado de acesso ordenando pelo estado com maior número (coluna address_state)?

```python
query = '''SELECT DISTINCT address_state, COUNT(*) AS users_by_state FROM (select cpf, address_state FROM `begrowth-user-api-demo.bg_users.bg_data_enginner_test_aisllan_max_caldeira`)  GROUP BY address_state ORDER BY users_by_state DESC'''
```

```python
pd.read_gbq(credentials=credentials, query=query)
```

# 4.2 - Quantos usuários únicos temos por estado de acesso ordenando pelo estado com maior número (coluna address_state)?

```python
query = '''SELECT DISTINCT address_state, COUNT(*) AS unique_users_by_state FROM (select distinct cpf, address_state FROM `begrowth-user-api-demo.bg_users.bg_data_enginner_test_aisllan_max_caldeira`)  GROUP BY address_state ORDER BY unique_users_by_state DESC'''
```

```python
pd.read_gbq(credentials=credentials, query=query)
```

# 4.3 - Quantos usuários temos por source ordando por ordem alfabética?

```python
query = '''SELECT utm_source, COUNT(*) AS users_by_source FROM `begrowth-user-api-demo.bg_users.bg_data_enginner_test_aisllan_max_caldeira` GROUP BY utm_source ORDER BY utm_source ASC'''
```

```python
pd.read_gbq(credentials=credentials, query=query)
```

# 4.4 - Quantos usuários únicos temos por source ordando por ordem alfabética?

```python
query = '''SELECT DISTINCT utm_source, COUNT(*) AS unique_users_by_source FROM (select distinct cpf, utm_source FROM `begrowth-user-api-demo.bg_users.bg_data_enginner_test_aisllan_max_caldeira`)  GROUP BY utm_source ORDER BY utm_source ASC'''
```

```python
pd.read_gbq(credentials=credentials, query=query)
```

# 4.5 - Sabendo disso, nosso time precisa saber qual a Região Fiscal que esse CPF foi registrado:

```python
query = '''SELECT C.uf As fiscal_region, D.*  FROM `begrowth-user-api-demo.bg_users.bg_data_enginner_test_aisllan_max_caldeira` D
INNER JOIN  `begrowth-user-api-demo.bg_users.cpf_state` C
              ON SUBSTR((CAST(D.cpf as string)), -3, 1) = CAST(C.id as string)'''
```

```python
pd.read_gbq(credentials=credentials, query=query)
```

# 4.6 - O estado de acesso obtido atraves das coordenadas geograficas fica na mesma Região Fiscal que esse CPF foi registrado? (Criar flag same_state no select)

```python
query = '''WITH A AS (SELECT C.uf As fiscal_region, D.address_state, D.* EXCEPT(address_state) FROM `begrowth-user-api-demo.bg_users.bg_data_enginner_test_aisllan_max_caldeira` D
LEFT JOIN  `begrowth-user-api-demo.bg_users.cpf_state` C
              ON SUBSTR((CAST(D.cpf as string)), -3, 1) = CAST(C.id as string)),
B AS (SELECT B.uf, initials FROM `begrowth-user-api-demo.bg_users.brazilian_state` B
LEFT JOIN `begrowth-user-api-demo.bg_users.cpf_state` C
  ON B.initials LIKE CONCAT ('%', C.uf, '%')
              OR C.uf LIKE CONCAT ('%', B.initials, '%'))

SELECT IF((B.initials LIKE CONCAT ('%', A.fiscal_region, '%') OR A.fiscal_region LIKE CONCAT ('%', B.initials, '%')), TRUE, FALSE) AS same_state, * EXCEPT(uf, initials) FROM A
LEFT JOIN B
  ON A.address_state = B.uf'''
```

```python
pd.read_gbq(credentials=credentials, query=query)
```
